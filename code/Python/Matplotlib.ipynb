```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('Y Axis Label')
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
plt.ylabel('Y Axis Label')
plt.show()
```

```python
numbers = np.arange(0, 50, 1)
squared_numbers = [number ** 2 for number in numbers]
plt.plot(numbers,squared_numbers)
plt.ylabel('Square of X')
plt.show()
```

## Scatter Plot

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
x1 = np.arange(100)
y1 = [number + 10 + random.randint(0,200)   for number in x1]
  
plt.scatter(x1, y1, c ="red", 
            linewidths = 2, 
            marker ="o", 
            s = 75)
  
  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
